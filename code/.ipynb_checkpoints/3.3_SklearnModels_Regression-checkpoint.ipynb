# Functions

```python
def MSE(y_true,y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return mse

def R2(y_true,y_pred):    
    r2 = r2_score(y_true, y_pred)
    return r2

def two_score(y_true,y_pred):    
    mse = MSE(y_true,y_pred) #set score here and not below if using MSE in GridCV
    score = R2(y_true,y_pred)
    return mse

def two_scorer():
    return make_scorer(two_score, greater_is_better=True)
```

```python
# Multicollinearity
def multicollinearity(model):
    for i in model.columns:
        for j in model.columns:
            if i != j:
                corr = model[i].corr(model[j])
                if corr > 0.7:
                    print('{} and {} corr = {}'.format(i,j,corr))
```

```python
# GridSearch Function
def pipeline_gridsearch(pipe, pipe_params, X_train, y_train):
    tscv = TimeSeriesSplit(n_splits = 5) # Time Series Split
    gs = GridSearchCV(pipe, param_grid=pipe_params, cv=tscv, n_jobs = -1, scoring=two_scorer())
    gs.fit(X_train, y_train)
    print('Best Score: ' + str(gs.best_score_)) # best_score_ is gridsearch's function
    print("Best Parameters: " + str(gs.best_params_))# best_params_ is gridsearch's function
```

# Feature Selection

We will use aapl to create our models.

I want to do a multi-step forecast of 3 periods, using data from 3, 4, 5 periods ago to predict. Therefore i will lag the columns by 3.

```python
# Creating 3 empty Columns to lag
aapl_lag = aapl.reindex(aapl.index.tolist() + list(range(len(aapl), len(aapl)+10)))
aapl = aapl.reindex(aapl.index.tolist() + list(range(len(aapl), len(aapl)+10)))
```

```python
# Create lag dataframe
appl_lag = aapl_lag.shift(periods = 10)
aapl_lag.shape
```

```python
# Merge shifted data with original close_price
aapl_lag = appl_lag.merge(aapl[['close_price', 'reversal']], how = 'left', on = aapl_lag.index).drop(columns = 'key_0')
# Drop first 3 rows
aapl_lag = aapl_lag.iloc[10:]
```

```python
aapl_lag
```

## Method 1 (Correlation)

```python
corr = aapl_lag.corr()
corr.sort_values(['close_price_y'], ascending = False, inplace = True)
corr['close_price_y'].head(54)
```

Since the top 54 features are very correlated with the close_price, we will use them in as our features. The features we will use is from others_cr to volatility_dchi.

### Target(y) and Features(X)

```python
X_corr = aapl_lag[corr['close_price_y'].iloc[0:54].index].drop(columns = ["close_price_y",'time']).drop(aapl_lag[corr['close_price_y'].iloc[0:54].index].drop(columns = ["close_price_y",'time']).tail(10).index)
X_corr_extra = aapl_lag[corr['close_price_y'].iloc[0:54].index].drop(columns = ["close_price_y",'time']).tail(10) # Last 3 rows to be combined with test set after train test split
y_corr = aapl_lag['close_price_y'].dropna()
```

```python
X_corr.shape , y_corr.shape , X_corr_extra.shape
```

## Method 2 (Lasso)

```python
features = aapl_lag.drop(columns = ['close_price_y', 'local_time', 'time','stock']).columns
X_lasso = aapl_lag[features].dropna()
y_lasso = aapl_lag['close_price_y'].dropna()
```

```python
ss = StandardScaler()
tscv = TimeSeriesSplit(n_splits = 5) # Time Series Split
lasso = LassoCV(n_alphas = 200, cv = tscv, n_jobs = -1)
```

```python
X_lasso = ss.fit_transform(X_lasso)
lasso.fit(X_lasso,y_lasso)
```

```python
lasso_coefs = pd.DataFrame({"variable": features,
                           "coef": lasso.coef_,
                           "abs_coef": np.abs(lasso.coef_)})
lasso_coefs.sort_values('abs_coef', inplace = True, ascending = False)
lasso_coefs.head(26)
```

As we can see from the coefficients that affect the close price. Variables from open_price to trend_vortex_ind_pos has some effect on the price. The other features got eliminated by the LassoCV

### Target(y) and Features(X)

```python
X_lasso = aapl_lag[lasso_coefs['variable'].iloc[0:26]].drop(aapl_lag[lasso_coefs['variable'].iloc[0:26]].tail(10).index)
X_lasso_extra = aapl_lag[lasso_coefs['variable'].iloc[0:26]].tail(10)
y_lasso = aapl_lag['close_price_y'].dropna()
```

```python
X_lasso.shape, y_lasso.shape, X_lasso_extra.shape
```

```python
X_corr_lags = aapl_lag[corr['close_price_y'].iloc[0:54].index].drop(columns = ["close_price_y",'time'])
X_corr_lags_1 = X_corr_lags.shift(8)
X_corr_lags_2 = X_corr_lags.shift(10)
```

```python
X_corr = X_corr_lags.merge(X_corr_lags_1, how = 'left', on = X_corr_lags.index).drop(columns = 'key_0')
X_corr_merged = X_corr.merge(X_corr_lags_2, how = 'left', on = X_corr.index).drop(columns = 'key_0')
```

```python
# Drop Top 2 rows because of NA
X_corr = X_corr_merged.iloc[10:].drop(X_corr_merged.tail(10).index)
X_corr_extra = X_corr_merged.tail(10) # Last 3 rows to be combined with test set after train test split
y_corr = y_corr.iloc[10:]
y_corr.shape, X_corr.shape, X_corr_extra.shape
```

### Lasso

```python
X_lasso_lags = aapl_lag[lasso_coefs['variable'].iloc[0:26]]
X_lasso_lags_1 = X_lasso_lags.shift(8)
X_lasso_lags_2 = X_lasso_lags.shift(10)
```

```python
X_lasso = X_lasso_lags.merge(X_lasso_lags_1, how = 'left', on = X_lasso_lags.index).drop(columns = 'key_0')
X_lasso_merged = X_lasso.merge(X_lasso_lags_2, how = 'left', on = X_lasso.index).drop(columns = 'key_0')
```

```python
# Drop Top 2 rows because of NA
X_lasso = X_lasso_merged.iloc[10:].drop(X_lasso_merged.tail(10).index)
X_lasso_extra = X_lasso_merged.tail(10) # Last 3 rows to be combined with test set after train test split
y_lasso = y_lasso.iloc[10:]
y_lasso.shape, X_lasso.shape
```

# Linear Regression

## Correlation

```python
# Create Pipeline
# LinearRegression
pipe_lr = Pipeline([
    ('ss', StandardScaler()),
    ('lr', LinearRegression())
])
# Lasso Regression
pipe_lasso = Pipeline([
    ('ss', StandardScaler()),
    ('lasso', LassoCV(n_alphas = 200))
])
# Ridge Regression
pipe_ridge = Pipeline([
    ('ss', StandardScaler()),
    ('ridge', RidgeCV(alphas = np.logspace(.1,10,100)))
])
```

```python
# Cross Validation
lr_score_c = cross_val_score(pipe_lr, X_training_c, y_training_c, cv=tscv)
print(f'r2 Score for Linear Regression: {lr_score_c.mean()}')
lr_score_c = cross_val_score(pipe_lasso, X_training_c, y_training_c, cv=tscv)
print(f'r2 Score for Lasso: {lr_score_c.mean()}')
lr_score_c = cross_val_score(pipe_ridge, X_training_c, y_training_c, cv=tscv)
print(f'r2 Score for Ridge: {lr_score_c.mean()}')
```

```python
# Fitting Model to Training Data
# Normalizing Training Data
ss = StandardScaler()
X_training_sc = ss.fit_transform(X_training_c)
X_val_sc = ss.transform(X_val_c)

# Linear Regression
lr = LinearRegression()
lr.fit(X_training_sc, y_training_c)

# r2 Score of training and val set
print(f'Training r2 Score(Linear): {lr.score(X_training_sc, y_training_c)}')
print(f'Validation r2 Score(Linear): {lr.score(X_val_sc, y_val_c)}')


prediction_lr_c = lr.predict(X_val_sc)
rmse_lr_c = np.sqrt(mean_squared_error(y_val_c, prediction_lr_c))
print(f'RMSE(Linear): {rmse_lr_c}')
print('-'*40)

# Lasso
lasso = LassoCV(n_alphas = 200)
lasso.fit(X_training_sc, y_training_c)

# r2 Score of training and val set
print(f'Training r2 Score(Lasso): {lasso.score(X_training_sc, y_training_c)}')
print(f'Validation r2 Score(Lasso): {lasso.score(X_val_sc, y_val_c)}')


prediction_lasso_c = lasso.predict(X_val_sc)
rmse_lasso_c = np.sqrt(mean_squared_error(y_val_c, prediction_lasso_c))
print(f'RMSE(Lasso): {rmse_lasso_c}')
print('-'*40)

# Ridge
ridge = RidgeCV(alphas = np.logspace(.1,10,100))
ridge.fit(X_training_sc, y_training_c)

# r2 Score of training and val set
print(f'Training r2 Score(Ridge): {ridge.score(X_training_sc, y_training_c)}')
print(f'Validation r2 Score(Ridge): {ridge.score(X_val_sc, y_val_c)}')


prediction_ridge_c = ridge.predict(X_val_sc)
rmse_ridge_c = np.sqrt(mean_squared_error(y_val_c, prediction_ridge_c))
print(f'RMSE(Ridge): {rmse_ridge_c}')
```

```python
# Create Pipeline
pipe_svr = Pipeline([
    ('ss', StandardScaler()),
    ('svr', SVR())
])
pipe_params_svr = {
    'svr__kernel': ['rbf', 'poly', 'sigmoid', 'linear'],
    'svr__gamma': ['auto', 0.1, 1],
    'svr__tol': [1,2,3],
    'svr__epsilon': [0.1,0.5,1]
}
```

```python
# GridSearch
pipeline_gridsearch(pipe_svr, pipe_params_svr, X_training_l,y_training_l)
```

```python
# Fitting Model to Training Data
# Normalizing Training Data
ss = StandardScaler()
X_training_sc = ss.fit_transform(X_training_l)
X_val_sc = ss.transform(X_val_l)

# RandomForest Regression
svr = SVR(epsilon = 0.1, gamma = 1, kernel = 'poly', tol = 1)
svr.fit(X_training_sc,y_training_l)

# r2 Score of training and val set
print(f'Training r2 Score(SVM): {svr.score(X_training_sc, y_training_l)}')
print(f'Validation r2 Score(SVM): {svr.score(X_val_sc, y_val_l)}')

prediction_svr_l = svr.predict(X_val_sc)
rmse_svr_l = np.sqrt(mean_squared_error(y_val_l, prediction_svr_l))
print(f'RMSE(SVM): {rmse_svr_l}')
```

## Plot Line Plot to see how closely predicted price follows

```python
# Create DataFrame to plot
model = pd.DataFrame(columns = ['val_price', 'predicted_price'])
model['val_price'] = y_val_l
model['predicted_price']= prediction_svr_l
model.sort_index(inplace=True)
```

```python
model_plot = model.iloc[:100]
```

```python
# Plot a lineplot between test data saleprice against predicted saleprice
plt.figure(figsize=(16,9))
plt.plot(model_plot.index, 'val_price', data = model_plot, marker='o', color='skyblue')
plt.plot(model_plot.index, 'predicted_price', data = model_plot, marker='x', color='olive')
plt.title(label = "Comparison Between Val Price and Predicted Price", fontsize = 16)
plt.legend(prop={'size':30})
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel("Date", fontsize = 16)
plt.ylabel("Stock Price", fontsize = 16)
```

Using the SVM Regression method, the priced price seems to be able to follow the actual price decently well. However, like using the linear Regression Method, there seems to be a lag in how the price is being predicted.
